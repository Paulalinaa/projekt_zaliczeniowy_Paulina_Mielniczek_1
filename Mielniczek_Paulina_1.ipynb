```python
import pandas as pd
df_zadanie = pd.read_csv("/dane1.csv")
df_zadanie
```

```python
#Zadanie 1a
mapa_dni = {
    'Mon': 'Poniedzia≈Çek',
    'Tue': 'Wtorek',
    'Wed': '≈öroda',
    'Thu': 'Czwartek',
    'Fri': 'PiƒÖtek',
    'Sat': 'Sobota',
    'Sun': 'Niedziela'
}
df_zadanie['created_at_polskie'] = df_zadanie['created_at'].str.replace(r'\b(?:Mon|Tue|Wed|Thu|Fri|Sat|Sun)\b', lambda x: mapa_dni[x.group()], regex=True)
print(df_zadanie["created_at_polskie"])
```

```python
#zadanie 1b
mapa_miesiecy = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
}
df_zadanie['created_at_polskie'] = df_zadanie['created_at_polskie'].str.replace(r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b', lambda x: mapa_miesiecy[x.group()], regex=True)

print(df_zadanie["created_at_polskie"])
```

```python
#zadanie 2g
dzien_tygodnia = df_zadanie['created_at'].str[:3].map(mapa_dni).value_counts()
dzien_tygodnia_slownik = dzien_tygodnia.to_dict()
print(dzien_tygodnia_slownik)
```

```python
# Zadanie 3a
!pip install spacy
!python -m spacy download pl_core_news_sm
import spacy
nlp = spacy.load("pl_core_news_sm")
```

```python
#zadanie 3b
def detect_pers_names(text):
    doc = nlp(text)
    pers_names = []
    for ent in doc.ents:
        if ent.label_ == "persName":
            pers_names.append(ent.text)
    return pers_names

df_zadanie['persons'] = df_zadanie['text'].apply(detect_pers_names)
print(df_zadanie[df_zadanie['persons'].apply(len) > 0]['persons'])
```

```python
#zadanie 3c
def detect_place_names(text):
    doc = nlp(text)
    place_names = []
    for ent in doc.ents:
        if ent.label_ == "placeName":
            place_names.append(ent.text)
    return place_names

df_zadanie['organizations'] = df_zadanie['text'].apply(detect_place_names)
print(df_zadanie[df_zadanie['organizations'].apply(len) > 0]['organizations'])
```

```python
#zadanie 3d
def detect_org_names(text):
    doc = nlp(text)
    org_names = []
    for ent in doc.ents:
        if ent.label_ == "orgName":
            org_names.append(ent.text)
    return org_names

df_zadanie['places'] = df_zadanie['text'].apply(detect_org_names)
print(df_zadanie[df_zadanie['places'].apply(len) > 0]['places'])
```
